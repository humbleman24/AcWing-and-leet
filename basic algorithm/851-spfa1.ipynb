### 851 SPFA1
#### Problem Description

Given a directed graph with \( n \) vertices and \( m \) edges, the graph may contain parallel edges and self-loops, and edge weights may be negative.

Please find the shortest distance from vertex 1 to vertex \( n \) using at most \( k \) edges. If it is not possible to reach vertex \( n \) from vertex 1, output `impossible`.

Note: The graph may contain negative weight cycles.


Input Format:

The first line contains three integers \( n \), \( m \), and \( k \).

The following \( m \) lines each contain three integers \( x \), \( y \), and \( z \), indicating that there is a directed edge from vertex \( x \) to vertex \( y \) with a length of \( z \).

Vertices are numbered from 1 to \( n \).


Output Format:

Output an integer, representing the shortest distance from vertex 1 to vertex \( n \) using at most \( k \) edges.

If no path meeting the conditions exists, output `impossible`.



Constraints:

1 ≤ n, m ≤ 10^5

The absolute value of any edge weight does not exceed 10,000.



Sample Input:

3 3

1 2 5

2 3 -3

1 3 4

Sample Output:

2

```python
from collections import deque
N,M = 100010, 100010
h,e,ne,w = [-1]*N, [-1]*N,[-1]*N,[0] * N
idx = 0
dist = [float('inf')] * N
dist[1] = 0

def add(a,b,z):
    global idx
    e[idx] = b
    ne[idx] = h[a]
    w[idx] = z
    h[a] = idx
    idx += 1

def spfa():
    visit = set()
    q = deque()
    q.append(1)
    visit.add(1)
    for i in range(n):
        while len(q):
            point = q.pop()
            visit.remove(point)
            next = h[point]
            while next != -1:
                j = e[next]
                if dist[j] > dist[point] + w[next]:
                    dist[j] = dist[point] + w[next]
                    if j not in visit:
                        q.append(j)
                        visit.add(j)
                next = ne[next]
    if dist[n] == float('inf'):
        return "impossible"
    return dist[n]

n,m = map(int,input().split())
for i in range(m):
    a,b,z = map(int,input().split())
    add(a,b,z)

print(spfa())
```

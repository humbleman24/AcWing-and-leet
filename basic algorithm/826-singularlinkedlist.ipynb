### 826 singular linked-list
#### Problem Description
Initially, the linked list is empty. The linked list should support three types of operations:

Insert a number at the head of the linked list.
Delete the number that is immediately after the k-th inserted number.
Insert a number immediately after the k-th inserted number.
You are required to perform these operations on the linked list a total of M times. After completing all operations, output the entire linked list from head to tail.

**Note:**

In the problem statement, the "k-th inserted number" does not refer to the k-th number in the current state of the linked list. Instead, it refers to the order in which numbers were inserted during the process. For instance, if n numbers have been inserted throughout the operations, these numbers are sequentially referred to as: the 1st inserted number, the 2nd inserted number, ..., and the nth inserted number.

Input Format:

The first line contains an integer M, which indicates the number of operations to be performed.

The following M lines each contain an operation command. The operation commands can be one of the following types:

- `H x`, which means insert a number x at the head of the linked list.
- `D k`, which means delete the number that follows the k-th inserted number. If k is 0, it means to delete the head node.
- `I k x`, which means insert a number x immediately after the k-th inserted number (in this operation, k is always greater than 0).


Output Format:

One line, containing linked list



Constraints: 
1 ≤ `M` ≤ 100000,

Sample Input:

10

H 9

I 1 1

D 1

D 0

H 6

I 3 6

I 4 5

I 4 5

I 3 4

D 6

Sample Output:

6 4 6 5

```python
class Linkedlist():
    class Node():
        def __init__(self,value,index):
            self.value = value
            self.index = index
            self.next = None

    def __init__(self):
        self.head = None
        self.index = 1
        self.length = 0
    
    def addhead(self,value):
        if self.head:
            newNode = self.Node(value,self.index)
            newNode.next = self.head
            self.head = newNode
        else:
            newNode = self.Node(value,self.index)
            self.head = newNode
        self.index += 1
        self.length += 1

    def insert(self,index,value):
        i = 0
        current = self.head
        while i < self.length:
            if current.index == index:
                break
            else:
                current = current.next
                i += 1
        newNode = self.Node(value,self.index)
        newNode.next = current.next
        current.next = newNode
        self.index += 1
        self.length += 1

    def delete(self,index):
        i = 0
        current = self.head
        while i < self.length:
            if index == 0:
                self.head = self.head.next
                break
            elif current.index == index:
                current.next = current.next.next
                break
            else:
                current = current.next
                i += 1

        self.length -= 1


    def traversal(self):
        node = self.head
        for i in range(self.length):
            print(node.value,end=' ')
            node = node.next
```

```python
M = int(input())
ll = Linkedlist()
for i in range(M):
    operant = input().split()
    if operant[0] == 'H':
        ll.addhead(operant[1])
    elif operant[0] == 'I':
        ll.insert(int(operant[1]),operant[2])
    elif operant[0] == 'D':
        ll.delete(int(operant[1]))
ll.traversal()
```

```python
# using the array to simulate the linked list, with much lower time complexity

N = 100010
e = [0] * N
ne = [0] * N
idx = 0
head = -1

def add_head(val):
    global head, idx
    e[idx] = val
    ne[idx] = head
    head = idx
    idx += 1

def delete_node(i):
    global head, idx
    if i == -1:
        head = ne[head]
    else:
        ne[i] = ne[ne[i]]

def insert_node(i,val):
    global head, idx
    e[idx] = val
    ne[idx] = ne[i]
    ne[i] = idx
    idx += 1

M = int(input())
for i in range(M):
    operant = input().split()
    if operant[0] == 'H':
        add_head(operant[1])
    elif operant[0] == 'I':
        insert_node(int(operant[1])-1,operant[2])
    elif operant[0] == 'D':
        delete_node(int(operant[1])-1)

current = head
while current != -1:
    print(e[current],end=' ')
    current = ne[current]
```

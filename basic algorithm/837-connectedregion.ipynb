### 837 connected components
#### Problem Description
Given an undirected graph with \( n \) vertices (numbered from 1 to \( n \)), initially, there are no edges in the graph.

Now, you need to perform \( m \) operations, and there are three types of operations:

- `C a b`, add an edge between vertex \( a \) and vertex \( b \). \( a \) and \( b \) may be the same;
- `Q1 a b`, query whether vertex \( a \) and vertex \( b \) are in the same connected component. \( a \) and \( b \) may be the same;
- `Q2 a`, query the number of vertices in the connected component containing vertex \( a \).


Input Format:

The first line inputs two integers \( n \) and \( m \).

The following \( m \) lines each contain an operation instruction, which is either `C a b`, `Q1 a b`, or `Q2 a`.


Output Format

For each query command `Q1 a b`, output a result. If \( a \) and \( b \) are in the same connected component, output `Yes`; otherwise, output `No`.

For each query command `Q2 a`, output an integer representing the number of vertices in the connected component containing vertex \( a \).

Each result should be on a new line.


Constraints

1  n, m â‰¤ 10^5


Sample Input:

5 5

C 1 2

Q1 1 2

Q2 1

C 2 5

Q2 5

Sample Output:

Yes

2

3

```python
def find(x):
    while graph[x] != x:
        graph[x] = graph[graph[x]]
        x = graph[x]
    return graph[x]

n, m = list(map(int,input().split()))
graph = [i for i in range(n+1)]
size = [1] * (n+1)

for i in range(m):
    command = input().split()
    if command[0] == 'C':
        a,b = int(command[1]), int(command[2])
        if find(a) != find(b):                        # must use find !!!!!!!
            size[find(b)] += size[find(a)]
            graph[find(a)] = find(b)
    elif command[0] == 'Q1':
        if find(int(command[1])) == find(int(command[2])):
            print('Yes')
        else:
            print('No')
    elif command[0] == 'Q2':
        print(size[find(int(command[1]))])
```

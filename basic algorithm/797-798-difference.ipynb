### 797 Difference

#### Problem Description
Input an integer sequence of length `n`.

Then input `m` operations, each operation contains three integers `l`, `r`, `c`, indicating to add `c` to every number in the sequence between `[l, r]`.

Please output the sequence after completing all operations.


Input Format:

The first line contains two integers `n` and `m`.

The second line contains `n` integers, representing the integer sequence.

The following `m` lines, each contain three integers `l`, `r`, `c`, representing an operation.



Output Format:

One line, containing `n` integers represent the final sequence.


Constraints: 
1 ≤ `n`, `m` ≤ 100000,
1 ≤ `l` ≤ `r` ≤ `n`,
-1000 ≤ `c` ≤ 1000,
-1000 ≤ Value of elements in the integer sequence ≤ 1000



Sample Input:

6 3

1 2 2 1 2 1

1 3 1

3 5 1

1 6 1

Sample Output:

3 4 5 3 4 2

```python
# # One dimension
n,m = list(map(int,input().split()))
a = list(map(int,input().split()))
# construct the difference matrix
b = [0]*n
for i in range (n-1):
    b[i] += a[i]
    b[i+1] -= a[i]
b[-1] += a[-1]
# b = [a[0]]
# b += [a[i+1] - a[i] for i in range(0,n-1)]
i = 0
while i < m:
    i += 1
    l,r,c = list(map(int,input().split()))
    b[l-1] += c
    if r == n:
        continue
    b[r] -= c
out = [0]
for i in range(n):
    out.append(out[-1]+b[i])
    print(out[-1],end=' ')
```

### 798 Difference of matrix

#### Problem Description
Input an `n` by `m` matrix of integers, then input `q` operations, each operation contains five integers `x1, y1, x2, y2, c`, where `(x1, y1)` and `(x2, y2)` represent the coordinates of the top-left and bottom-right corners of a submatrix, respectively.

Each operation should add `c` to the value of each element within the selected submatrix.

Please output the matrix after completing all operations.


Input Format:

The first line contains integers `n`, `m`, `q`.

The next `n` lines, each contain `m` integers, representing the integer matrix.

The following `q` lines, each contain 5 integers `x1, y1, x2, y2, c`, representing an operation.




Output Format:

There are `n` lines in total, each with `m` integers, representing the final matrix after all operations have been completed.



Constraints: 
1 ≤ `n`, `m` ≤ 1000,
1 ≤ `q` ≤ 100000,
1 ≤ `x1` ≤ `x2` ≤ `n`,
1 ≤ `y1` ≤ `y2` ≤ `m`,
-1000 ≤ `c` ≤ 1000,
-1000 ≤ Value of elements within the matrix ≤ 1000




Sample Input:

3 4 3

1 2 2 1

3 2 2 1

1 1 1 1

1 1 2 2 1

1 3 2 3 2

3 1 3 4 1

Sample Output:

2 3 4 1

4 3 4 1

2 2 2 2

```python
# two dimension
n,m,q = list(map(int,input().split()))
b = [[0 for i in range(m)] for j in range(n)]
for i in range(n-1):
    col = list(map(int,input().split()))
    for j in range(m-1):
        b[i][j] += col[j]
        b[i][j+1] -= col[j]
        b[i+1][j] -= col[j]
        b[i+1][j+1] += col[j]
    b[i][-1] += col[-1]
    b[i+1][-1] -= col[-1]
col = list(map(int,input().split()))
for j in range(m-1):
    b[-1][j] += col[j]
    b[-1][j+1] -= col[j]
b[-1][-1] += col[-1]    

i = 0
while i < q:
    i += 1
    x1,y1,x2,y2,c = list(map(int,input().split()))
    x1,y1,x2,y2 = x1-1,y1-1,x2-1,y2-1
    b[x1][y1] += c
    if y2 < m-1:
        b[x1][y2+1] -= c                            # +1 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    if x2 < n-1:
        b[x2+1][y1] -= c                            # +1 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    if y2 < m-1 and x2 < n-1:
        b[x2+1][y2+1] += c                          # +1 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

a = [[0 for i in range(m+1)] for j in range(n+1)]
for i in range(1,n+1):
    for j in range(1,m+1):
        a[i][j] = a[i][j-1] + a[i-1][j] - a[i-1][j-1] + b[i-1][j-1]
        print(a[i][j], end=' ')
    print()
```

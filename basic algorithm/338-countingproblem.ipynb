### 338 counting problem
#### Problem Description
Given two integers \( a \) and \( b \), calculate the occurrence of each digit from 0 to 9 in all numbers between \( a \) and \( b \) inclusive.

For example, if \( a = 1024 \) and \( b = 1032 \), there are 9 numbers between \( a \) and \( b \) as follows:

1024, 1025, 1026, 1027, 1028, 1029, 1030, 1031, 1032

Among these numbers, the digit 0 appears 10 times, 1 appears 10 times, 2 appears 7 times, 3 appears 3 times, and so onâ€¦


Input Format:

Input contains multiple test cases.

Each test case is on a single line containing two integers \( a \) and \( b \).

Input ends when a line containing "0 0" is read, and this line should not be processed.


Output Format

For each test case, output a result on a single line.

Each result should contain ten numbers separated by spaces, where the first number indicates the number of times the digit 0 appears, the second number indicates the number of times the digit 1 appears, and so on up to the digit 9.


Constraints

0< a , b < 100000000

```python


def get_pre(nums,h,l):
    num = 0
    for i in range(h,l,-1):                        # not include l
        num = num * 10 + nums[i]
    return num


def count(num, u):
    if not num: return 0
    nums = list()
    while num:                         # convert into list, highest position at the back of list
        nums.append(num%10)
        num //= 10
    n = len(nums)
    res = 0
    for i in range(n-1,-1,-1):
        if i < n - 1:                  # not the highest position
            if u:                      # u is not 0
                res += get_pre(nums,n-1,i) * (10 ** i)                     # highest * 10 ^ i
            else:
                res += (get_pre(nums,n-1,i)-1) * (10 ** i)                 # highest - 1     when u == 0  
        if i == n - 1 and not u:          # when i is the highest and u == 0, not need to consider anymore
            continue
        if nums[i] == u:
            res += get_pre(nums, i - 1, -1) + 1             # not include i but include 0!!! from 00000 !!
        elif nums[i] > u:
            res += 10 ** i
    return res



a,b = map(int,input().split())
if a > b: a, b = b, a

while a or b:
    for i in range(10):
        print(count(b,i) - count(a-1,i), end=' ')
    print()
        
    a,b = map(int,input().split())
    if a > b: a, b = b, a
```

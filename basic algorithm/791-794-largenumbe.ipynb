### 791 addition

#### Problem Description
Given two positive integers(without prefix 0), calculate the addition of them


Input Format:

Two lines, with one integer in each line

Output Format:

One line, containing addition.


Constraints: 1 ≤ Integer.length ≤ 100000

Sample Input:

12

23

Sample Output:

35

```python
def add(ln,lm):
    out = list()
    t = 0
    i = 0
    while i < len(ln) or i < len(lm):
        if i < len(ln): t += ln[i]
        if i < len(lm): t += lm[i]
        out.append(t%10)
        t = t//10
        i += 1
    if t: out.append(t)
    return out
```

### 792 Subtraction

#### Problem Description
Given two positive integers(without prefix 0), calculate the subtraction of them


Input Format:

Two lines, with one integer in each line

Output Format:

One line, containing subtraction.


Constraints: 1 ≤ Integer.length ≤ 100000

Sample Input:

12

23

Sample Output:

-11

```python
def subtraction(ln,lm):
    reverse = False           
    # we need to gurantee that ln is larger than lm, so that we can use the bit-wise substraction
    if len(ln) < len(lm):
        ln, lm = lm, ln
        reverse = True
    elif len(ln) == len(lm):
        i = len(ln) - 1
        while i >= 0:
            if ln[i] > lm[i]:
                break
            elif ln[i] < lm[i]:
                ln, lm= lm, ln
                reverse = True
                break
            i -= 1

    out = list()
    i = 0
    while i < len(ln) or i < len(lm):
        if i >= len(lm):
            out.append(ln[i])
        else:
            if ln[i] - lm[i] < 0:               # sometimes forget this
                j = i + 1                       # start with the next bit!
                while ln[j] <= 0:               # keep each bit non-negative! would be a problem when there is no need to substraction, there is no chance to brown 10!
                    ln[j] += 9
                    j += 1
                ln[j] -= 1              
                out.append(10 + ln[i] - lm[i])
            else:
                out.append(ln[i] - lm[i])
        i += 1
    while len(out) > 1 and out[-1] == 0:         # prefix 0s need to be removed
        out.pop(-1)
        
    if reverse:
        out.append('-')
    return out
```

### 793 Multiply

#### Problem Description
Given two non-negative integers(without prefix 0), calculate the product of them


Input Format:

Two lines, with one integer in each line

Output Format:

One line, containing product.


Constraints: 1 ≤ first_Integer.length ≤ 100000, 0 ≤ second_Integer ≤ 100000

Sample Input:

2

3

Sample Output:

6

```python
def multilpy(ln,b):
    out = list()
    t = 0
    for i in range(len(ln)):
        t += b * ln[i]
        out.append(t%10)
        t //= 10
    while t != 0:
        out.append(t%10)
        t //= 10
    while len(out)>1 and out[-1] == 0:
        out.pop(-1)
    return out
```

### 794 divide

#### Problem Description
Given two non-negative integers (without leading zeros) `A` and `B`, calculate the quotient and remainder of `A` divided by `B`.

Input Format:

Two lines, with one integer in each line

Output Format:

Two lines, containing quotient and remainder each line.


Constraints: 1 ≤ first_Integer.length ≤ 100000, 0 ≤ second_Integer ≤ 100000

Sample Input:

7

2

Sample Output:

3

1

```python
def divide(ln,b):
    out = list()
    remain = 0
    for i in range(len(ln)-1,-1,-1):
        remain = remain*10 + ln[i]
        out.append(remain//b)
        remain = remain%b
    out.reverse()
    while len(out) > 1 and out[-1] == 0:
        out.pop(-1)
    return (out,remain)
```

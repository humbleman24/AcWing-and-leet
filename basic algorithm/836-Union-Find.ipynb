### 836 Union-Find
#### Problem Description
There are a total of \( n \) numbers, labeled from 1 to \( n \), initially each number is in its own set.

Now, you need to perform \( m \) operations, and there are two types of operations:

- `M a b`, merge the sets containing the numbers labeled \( a \) and \( b \). If the two numbers are already in the same set, ignore this operation;
- `Q a b`, query whether the numbers labeled \( a \) and \( b \) are in the same set.

Input Format:

The first line inputs two integers \( n \) and \( m \).

The following \( m \) lines each contain an operation instruction, which is either `M a b` or `Q a b`.



Output Format

For each query command `Q a b`, output a result. If \( a \) and \( b \) are in the same set, output `Yes`; otherwise, output `No`.

Each result should be on a new line.

Constraints

1  n, m â‰¤ 10^5


Sample Input:

4 5

M 1 2

M 3 4

Q 1 2

Q 1 3

Q 3 4

Sample Output:

Yes

No

Yes

```python
def find(x):
    while (x != union[x]):
        union[x] = union[union[x]]
        x = union[x]
    return union[x]


n,m = list(map(int,input().split()))
union = [i for i in range(1,n+1)]
for i in range(m):
    command = input().split()
    if command[0] == 'M':
        union[find(int(command[1]))] = find(int(command[2]))
    elif command[0] == 'Q':
        if find(int(command[1])) == find(int(command[2])):
            print('YES')
        else:
            print('NO')
```

### 861 Maximum match for bipartite graph
#### Problem Description

Given a bipartite graph where the left side contains \( n1 \) vertices (numbered 1 to \( n1 \)) and the right side contains \( n2 \) vertices (numbered 1 to \( n2 \)), the graph has a total of \( m \) edges.

It is guaranteed that any edge connects vertices from different parts.

Please determine the maximum matching of the bipartite graph.

**Bipartite Graph Matching:** In a given bipartite graph \( G \), if a subgraph \( M \) has an edge set {E} where no two edges share a vertex, then \( M \) is called a matching.

**Maximum Matching of a Bipartite Graph:** Among all possible matchings, the one with the most edges is called the maximum matching of the bipartite graph, and its number of edges is the maximum matching number.


Input Format:

Output an integer, representing the maximum number of matches in the bipartite graph.


Output Format:

If the given graph is a bipartite graph, output `Yes`. Otherwise, output `No`.

Constraints:

1≤n1,n2≤500,

1≤u≤n1,

1≤v≤n2,

1≤m≤10^5


Sample Input:

2 2 4

1 1

1 2

2 1

2 2

Sample Output:

2

```python
N = 500
M = 100010
h,e,ne,match = [-1] * N, [0] * M, [0] * M, [0] *N
idx = 0

def add(a,b):
    global idx
    e[idx] = b
    ne[idx] = h[a]
    h[a] = idx
    idx += 1

def find(index):
    next = h[index]
    while next != -1:
        j = e[next]
        if j not in visit:
            visit.add(j)
            if match[j] == 0 or find(match[j]):
                match[j] = index
                return True
        next = ne[next]
    return False
        


n1,n2, m = map(int,input().split())

for i in range(m):
    a,b = map(int,input().split())
    add(a,b)

res = 0
for i in range(1,n1 + 1):
    visit  = set()                             
# used inside of iteration to avoid searching the same point, but should be redefined before next generation because the nodes can be search in the next step
    if find(i):
        res += 1
print(res)
```

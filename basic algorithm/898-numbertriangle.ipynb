### 898 number triangle
#### Problem Description
Given a number triangle as shown below, start from the top. At each node, you can choose to move to the node directly below on the left or directly below on the right, continuing to the bottom layer. The goal is to find a path such that the sum of the numbers along the path is maximized.

                7

              3   8

            8   1   0

          2   7    4   0
          
       4    5    2   6    5



Input Format:

The first line contains an integer \( n \), representing the number of layers in the number triangle.

The following \( n \) lines each contain several integers, where the \( i \)-th line represents the integers contained in the \( i \)-th layer of the number triangle.



Output Format

Output an integer, representing the maximum value achievable.


Constraints

1 ≤ n ≤ 500,

-10,000 ≤ Integers in the triangle ≤ 10,000



Sample Input:

5

7

3 8

8 1 0 

2 7 4 4

4 5 2 6 5

Sample Output:

30

```python
N = 510
f = [[0 for i in range(N)] for j in range(N)]
value = [[0 for i in range(N)] for j in range(N)]

n= int(input())
for i in range(1, n+1):
    arr = list(map(int,input().split()))
    for j in range(1,i+1):
        value[i][j] = arr[j-1]

        
for i in range(n,0,-1):
    for j in range(1,i+1):
        f[i][j] = max(f[i+1][j], f[i+1][j+1])
        f[i][j] += value[i][j]

print(f[1][1])






```

### 803 merge interval

#### Problem Description
Given `n` intervals `[li, ri]`, merge all overlapping intervals.

Note that intervals that intersect at the endpoints are also considered overlapping.

Output the number of intervals after merging.

For example: `[1,3]` and `[2,6]` can be merged into one interval `[1,6]`.



Input Format:

The first line contains an integer `n`.

The next `n` lines, each contain two integers `l` and `r`.



Output Format:

One line, containing a integer, representing the intervals number after merge the overlapped interval



Constraints: 
1 ≤ `n` ≤ 100000,
10^9 ≤ `li` ≤ `ri` ≤ 10^9

Sample Input:

5

1 2

2 4

5 6

7 8

7 9

Sample Output:

3


```python
n = int(input())
a = list()
for i in range(n):
    b = list(map(int,input().split()))
    j = 0
    while j < len(a):                      # search for the place the add new interval
        if a[j][0] < b[0]:                 # order by left point   
            j += 1
        elif a[j][0] == b[0]:              # right point, when left points are equal
            if a[j][1] < b[1]:
                j += 1
            else:
                break
        else:                              # find the place
            break
    a.insert(j,b)

    k = 0
    while k < len(a)-1:                     # merge
        if a[k][1] >= a[k+1][0]:            # only to consider this
            if a[k][1] < a[k+1][1]:         # entirely cover
                a[k][1] = a[k+1][1]
            a.pop(k+1)
        else:
            k += 1

print(len(a))
```

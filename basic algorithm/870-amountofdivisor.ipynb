### 870 amount of divisor
#### Problem Description
Given \( n \) positive integers \( a_i \), please output the number of divisors of the product of these numbers, with the answer modulo \( 10^9 + 7 \).


Input Format:

The first line contains an integer \( n \).

The following \( n \) lines each contain one positive integer \( a_i \).



Output Format

Output an integer representing the number of divisors of the product of the given positive integers, with the result taken modulo \( 10^9 + 7 \).



Constraints

1 ≤ n ≤ 100,

2 ≤ a_i ≤ 2 * 10^9


Sample Input:

3

2

6

8


Sample Output:

12


```python
n = int(input())
primes = dict()
mod = 1e9 + 7

for i in range(n):
    x = int(input())
    i = 2
    while i <= x/i:
        if x % i == 0:
            if i not in primes.keys():       # do not initialize dict when it already have one!!!
                primes[i] = 0
            while x % i == 0:
                primes[i] += 1         
                x = int(x/i)
        i += 1
    x = int(x)
    if x > 1:
        if x not in primes.keys():       # do not initialize dict when it already have one!!!
            primes[x] = 1
        else:
            primes[x] += 1


res = 1
for value in primes.values():
    res *= (value + 1) % mod

print(res)
```

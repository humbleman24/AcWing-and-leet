### 2 Backpack
#### Problem Description
There are \( N \) items and a backpack with a capacity of \( V \). Each item can only be used once.

The volume of the \( i \)-th item is \( v_i \) and its value is \( w_i \).

Determine which items to put into the backpack so that the total volume of the items does not exceed the capacity of the backpack and the total value is maximized.
Output the maximum value.


Input Format:

The first line contains two integers, \( N \) and \( V \), separated by a space, representing the number of items and the capacity of the backpack, respectively.

The next \( N \) lines each contain two integers \( v_i \) and \( w_i \), separated by a space, representing the volume and value of the \( i \)-th item, respectively.



Output Format

Output an integer, representing the maximum value achievable.


Constraints

0 < N, V ≤ 1000

0 < v_i, w_i ≤ 1000


Sample Input:

4 5

1 2

2 4

3 4

4 5

Sample Output:

8

```python
# 2 D
N = 1010
f = [[0 for i in range(N)] for j in range(N)]
v, w = [0] * N, [0] * N

n,m = map(int,input().split())
for i in range(1,n+1):
    v[i], w[i] = map(int,input().split())

for i in range(1,n+1):
    for j in range(1,m+1):
        f[i][j] = f[i-1][j]
        if j >= v[i]:                         # equal can also be choosed
            f[i][j] = max(f[i][j], f[i-1][j-v[i]] + w[i])
print(f[n][m])

```

```python
# 1 D
N = 1010
f = [0 for i in range(N)]
v, w = [0] * N, [0] * N

n,m = map(int,input().split())
for i in range(1,n+1):
    v[i], w[i] = map(int,input().split())

for i in range(1,n+1):
    j = m
    while j >= v[i]:
        f[j] = max(f[j], f[j-v[i]] + w[i])
        j -= 1
print(f[m])

```

### 902 shortest edited distance
#### Problem Description
Given two strings, \( A \) and \( B \), you are to transform \( A \) into \( B \) using a series of operations. The available operations are:

- **Delete** – Remove a character from string \( A \).
- **Insert** – Insert a character at any position in string \( A \).
- **Replace** – Replace a character in string \( A \) with another character.

Determine the minimum number of operations required to transform \( A \) into \( B \).


Input Format:

The first line contains an integer \( n \), representing the length of string \( A \).

The second line contains a string of length \( n \), representing string \( A \).

The third line contains an integer \( m \), representing the length of string \( B \).

The fourth line contains a string of length \( m \), representing string \( B \).

Both strings contain only uppercase and lowercase letters.


Output Format

Output an integer, representing minist manipulation.


Constraints

1 ≤ n,m ≤ 1000

Sample Input:

10

AGTCTGACGC

11

AGTAAGTAGGC

Sample Output:

4

```python
N = 1010
f = [[0 for i in range(N)] for j in range(N)]
n = int(input())
a = [0] + list(input())
m = int(input())
b = [0] + list(input())

for i in range(m + 1):f[0][i] = i                # initialize!! take care of the range!
for i in range(n + 1):f[i][0] = i



for i in range(1, n+1):
    for j in range(1, m + 1):
        if a[i] != b[j]:
            f[i][j] = min(f[i-1][j], f[i][j-1], f[i-1][j-1])
            f[i][j] += 1
        else:                                         # !! important forget!
            f[i][j] = min(f[i-1][j-1], f[i-1][j] + 1, f[i][j-1] + 1)

print(f[n][m])
```

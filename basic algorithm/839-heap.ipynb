### 839 Heap 
#### Problem Description
Maintain a set that is initially empty, supporting the following operations:

- `I x`, insert a number \( x \);
- `PM`, output the current minimum value in the set;
- `DM`, delete the current minimum value in the set (it is guaranteed that this minimum value is unique);
- `D k`, delete the \( k \)-th inserted number;
- `C k x`, change the \( k \)-th inserted number to \( x \);

Now, you need to perform \( N \) operations. For all the second operations, output the current minimum value in the set.

Input Format:

The first line contains an integer \( N \).

The following \( N \) lines each contain an operation instruction, which can be one of the following: `I x`, `PM`, `DM`, `D k`, or `C k x`.

Output Format

For each `PM` output instruction, output a result representing the current minimum value in the set.

Each result should be on a new line.

Constraints

1 ≤ N ≤ 10^5

-10^9 ≤ x ≤ 10^9

Data is guaranteed to be valid.


Sample Input:

8

I -10

PM

I -10

D 1

C 2 8

I 6

PM

DM

Sample Output:

-10

6

```python
def down(x):
    t = x
    if 2*x <= size and h[t] > h[2*x]: t = 2 * x
    if 2*x +1 <= size and h[t] > h[2*x+1]: t = 2*x + 1
    if t != x:
        heap_swap(x,t)
        down[t]

def up(x):
    if x//2 > 0 and h[x] < h[x//2]:
        heap_swap(x,x//2)
        up(x//2)

def heap_swap(a,b):
    ph[hp[a]], ph[hp[b]] = ph[hp[b]], ph[hp[a]]
    hp[a], hp[b] = hp[b], hp[a]
    h[a], h[b] = h[b], h[a]



h = [0] * 100010
ph = [0] * 100010
hp = [0] * 100010
size = 0
idx = 0
n = int(input())

for i in range(n):
    command = input().split()
    if command[0] == 'I':
        size += 1
        h[size] = int(command[1])
        idx += 1
        ph[idx] = size
        hp[size] = idx
        up(size)
    elif command[0] == 'PM':
        print(h[1])
    elif command[0] == 'DM':
        heap_swap(1,size)
        size -= 1
        down(1)
    elif command[0] == 'D':
        k = ph[int(command[1])]
        heap_swap(k,size)
        size -= 1
        up(k)
        down(k)
    elif command[0] == 'C':
        k = ph[int(command[1])]
        h[k] = int(command[2])
        down(k)
        up(k)


```

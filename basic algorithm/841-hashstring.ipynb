### 841 hash string
#### Problem Description
Given a string of length \( n \), and \( m \) queries, each query contains four integers \( l1, r1, l2, r2 \). You need to determine whether the substrings of the string within the intervals \([l1, r1]\) and \([l2, r2]\) are exactly the same.

The string contains only uppercase and lowercase English letters and numbers.

Input Format:

The first line contains two integers \( n \) and \( m \), representing the length of the string and the number of queries.

The second line contains a string of length \( n \), which only includes uppercase and lowercase English letters and numbers.

The following \( m \) lines each contain four integers \( l1, r1, l2, r2 \), representing the two intervals involved in each query.

Note that the positions in the string start from 1.

Output Format

For each query, output a result. If the two substrings are exactly the same, output `Yes`; otherwise, output `No`.

Each result should be on a new line.


Constraints

1 ≤ n, m ≤ 10^5


Sample Input:

8 3

aabbaabb

1 3 5 7

1 3 6 8

1 2 1 2


Sample Output:

Yes

No

Yes

```python


N = 100010
P = 131
Q = 1 << 64
h = [0] * N
p = [1] * N

n,m = map(int,input().split())
arr = list(input())
for i in range(1,n+1):
    p[i] = P * p[i-1] % Q                        # easy to multiply!
    h[i] = (h[i-1]*P + ord(arr[i-1])) % Q        # 


for i in range(m):
    l1,r1,l2,r2 = map(int,input().split())
    a = (h[r1] - h[l1-1]*p[r1-l1+1]) % Q
    b = (h[r2] - h[l2-1]*p[r2-l2+1]) % Q
    if a == b:
        print('Yes')
    else:
        print('No')


```

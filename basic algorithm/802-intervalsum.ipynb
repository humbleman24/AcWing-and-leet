### 802 the sum of interval in infinite long array

#### Problem Description
Assume there is an infinitely long number axis, with every coordinate on the axis initially set to 0.

Now, we first perform `n` operations, each operation adds `c` to the number at position `x`.

Then, we conduct `m` queries, each query consists of two integers `l` and `r`, and you need to calculate the sum of all numbers in the interval `[l, r]`.


Input Format:

The first line contains two integers `n` and `m`.

The next `n` lines, each contain two integers `x` and `c`.

Then the following `m` lines, each contain two integers `l` and `r`.



Output Format:

There are `m` lines in total, each line outputs the sum of numbers within the interval queried.


Constraints: 
-10^9 ≤ `x` ≤ 10^9,
1 ≤ `n`, `m` ≤ 10^5,
-10^9 ≤ `l` ≤ `r` ≤ 10^9,
-10,000 ≤ `c` ≤ 10,000


Sample Input:

3 3

1 2

3 6

7 5

1 3

4 6

7 8

Sample Output:

8

0

5


```python
n,m = list(map(int,input().split()))
N = 300010

a = [0 for i in range(N)]
s = [0 for i in range(N)]
index = list()
add = list()
query = list()

# construct a index list, containing all the index would be used in addition and query
for i in range(n):
    x,c = list(map(int,input().split()))
    index.append(x)
    add.append((x,c))
for j in range(m):
    l,r = list(map(int,input().split()))
    index.append(l)
    index.append(r)
    query.append((l,r))

index.sort()
index = list(set(index))
```

```python
def find(x):                 # to find the real index, binary search
    l = 0
    r = len(index) - 1
    while l < r:
        mid = l + r >> 1
        if index[mid] >= x:
            r = mid
        else:
            l = mid + 1
    return r + 1
```

```python
for x,c in add:
    x = find(x)
    a[x]+=c

for i in range(1,len(index)+1):
    s[i] = s[i-1] + a[i]

for l,r in query:
    l = find(l)
    r = find(r)
    print(s[r] - s[l-1])
```

### 9 divided backpack
#### Problem Description
There are \( N \) groups of items and a backpack with a capacity of \( V \).

Each group contains several items, but only one item from each group can be chosen. Each item has a volume \( v_{ij} \) and a value \( w\_{ij} \), where \( i \) is the group number and \( j \) is the item number within the group.

Determine which items to put into the backpack so that the total volume of the items does not exceed the capacity of the backpack and the total value is maximized.

Output the maximum value.


Input Format:

The first line contains two integers \( N \) and \( V \), separated by a space, representing the number of item groups and the backpack capacity, respectively.

Each of the following \( N \) groups of data starts with:

- A line containing an integer \( S_i \), representing the number of items in the \( i \)-th group;
- Followed by \( S_i \) lines, each containing two integers \( v_{ij} \) and \( w_{ij} \), separated by a space, representing the volume and value of the \( j \)-th item in the \( i \)-th group.


Output Format

Output an integer, representing the maximum value achievable.


Constraints

0 < N, V ≤ 100

0 < S_i ≤ 100

0 < v_{ij}, w_{ij} ≤ 100


Sample Input:

3 5

2

1 2

2 4

1

3 4

1

4 5

Sample Output:

8

```python
N = 110
v, w, f = [[0 for i in range(N)] for j in range(N)], [[0 for i in range(N)] for j in range(N)],[[0 for i in range(N)] for j in range(N)]
s = [0] * N

n,m = map(int,input().split())
for i in range(1,n+1):
    g = int(input())
    s[i] = g
    for j in range(g):
        v[i][j], w[i][j] = map(int,input().split())

for i in range(1,n+1):
    for j in range(1, m+1):
        f[i][j] = f[i-1][j]
        for k in range(s[i]):
            if j >= v[i][k]:
                f[i][j] = max(f[i][j], f[i-1][j-v[i][k]] + w[i][k])

print(f[n][m])
```

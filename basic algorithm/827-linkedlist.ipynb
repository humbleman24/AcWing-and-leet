### 827 linked-list
#### Problem Description

Initially, the doubly linked list is empty. The doubly linked list should support five types of operations:

1. Insert a number at the leftmost position.
2. Insert a number at the rightmost position.
3. Delete the k-th inserted number.
4. Insert a number immediately to the left of the k-th inserted number.
5. Insert a number immediately to the right of the k-th inserted number.

You are required to perform these operations on the doubly linked list a total of M times. After completing all operations, output the entire linked list from left to right.

**Note:**

In the problem statement, the "k-th inserted number" does not refer to the k-th number in the current state of the linked list. Instead, it refers to the order in which numbers were inserted during the process. For instance, if n numbers have been inserted throughout the operations, these numbers are sequentially referred to as: the 1st inserted number, the 2nd inserted number, ..., and the nth inserted number.

Input Format:

The first line contains an integer M, which indicates the number of operations to be performed.

The following M lines each contain an operation command. The operation commands can be one of the following types:

- `L x`: Insert a number x at the leftmost position of the linked list.
- `R x`: Insert a number x at the rightmost position of the linked list.
- `D k`: Delete the k-th inserted number.
- `IL k x`: Insert a number x immediately to the left of the k-th inserted number.
- `IR k x`: Insert a number x immediately to the right of the k-th inserted number.


Output Format:

One line, containing linked list



Constraints: 
1 ≤ `M` ≤ 100000,

Sample Input:

10

R 7

D 1

L 3

IL 2 10

D 3

IL 2 7
 
L 8

R 9

IL 4 7

IR 2 2

Sample Output:

8 7 7 3 2 9

```python
N = 100010
e = [0] * N
left = [0] * N
right = [0] * N
head = -1
tail = -1
idx = 0

def add_l(val):
    global head,tail,idx
    if head != -1:
        e[idx] = val
        right[idx] = head
        left[idx] = -1
        left[head] = idx
        head = idx
        
    elif head == -1:
        e[idx] = val
        right[idx] = head
        left[idx] = -1
        head = idx
        tail = idx
    idx += 1

def add_r(val):
    global head,tail,idx
    if tail != -1:
        e[idx] = val
        left[idx] = tail
        right[idx] = -1
        right[tail] = idx
        tail = idx
    elif tail == -1:
        e[idx] = val
        left[idx] = tail
        right[idx] = -1
        tail = idx
        head = idx
    idx += 1

def delete_node(i):
    global head,tail,idx
    left[right[i]] = left[i]
    right[left[i]] = right[i]

    if head == i:
        head = left[head]
    if tail == i:
        tail = right[tail]

def insert_l(i,val):
    global head,tail,idx
    e[idx] = val
    right[idx] = i
    left[idx] = left[i]
    right[left[i]] = idx
    left[i] = idx
    if head == i:
        head = idx
    idx += 1

def insert_r(i,val):
    global head,tail,idx
    e[idx] = val
    right[idx] = right[i]
    left[idx] = i
    left[right[i]] = idx
    right[i] = idx
    if tail == i:
        tail = idx
    idx += 1

 
M = int(input())
for i in range(M):
    operant = input().split()
    if operant[0] == 'L':
        add_l(operant[1])
    elif operant[0] == 'R':
        add_r(operant[1])
    elif operant[0] == 'IL':
        insert_l(int(operant[1])-1,operant[2])
    elif operant[0] == 'IR':
        insert_r(int(operant[1])-1,operant[2])
    elif operant[0] == 'D':
        delete_node(int(operant[1])-1)

current = head
while current != -1:
    print(e[current],end=' ')
    current = right[current]
```

### 846 centroid of tree
#### Problem Description
Given a tree containing \( n \) nodes (numbered from 1 to \( n \)) and \( n-1 \) undirected edges.

Find the centroid of the tree and output the maximum number of nodes in the largest connected component after removing the centroid.

**Definition of Centroid**: A centroid is a node in a tree such that removing this node minimizes the maximum number of nodes in any of the remaining connected components. The node that achieves this is called the centroid of the tree.


Input Format:

The first line contains an integer \( n \), representing the number of nodes in the tree.

The next \( n-1 \) lines each contain two integers \( a \) and \( b \), indicating that there is an edge between node \( a \) and node \( b \).


Output Format

Output an integer \( m \), representing the maximum number of nodes in the largest connected component after the centroid is removed.



Constraints

1 ≤ n ≤ 10^5


Sample Input:

9

1 2

1 7

1 4

2 8

2 5

4 3

3 9

4 6


Sample Output:

4

```python
N = 100010
M = 2*N
# h is the start of an edge, e store the out point of an edge, ne implement the adjcent list used to find all the edges of start point
h, e, ne = [-1] * N, [-1]*M, [-1]*M
idx = 0

def add(a, b):
    global idx
    e[idx] = b
    ne[idx] = h[a]
    h[a] = idx
    idx += 1

def dfs(node):
    global ans
    sum, res = 1, 0
    visit.add(node)

    next_p = h[node]                # mistake in take the point
    # must choose the index first, because rely on this index to get the next index and the while also depends on index
    while next_p != -1:
        new_p = e[next_p]
        if new_p not in visit:
            s = dfs(new_p)
            res = max(res, s)        # store the biggest sub component
            sum += s                 # add wrong at first!
        next_p = ne[next_p]

    res = max(res, n - sum)          # use the sum! it is the sum of sub tree.

    ans = min(ans, res)
    return sum


n = int(input())
visit = set()
ans = N
for i in range(n-1):
    a,b = map(int,input().split())
    add(a,b)
    add(b,a)
dfs(1)
print(ans)
```

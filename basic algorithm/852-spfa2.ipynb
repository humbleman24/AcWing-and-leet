### 852 SPFA2
#### Problem Description

Given a directed graph with \( n \) vertices and \( m \) edges, the graph may contain parallel edges and self-loops, and edge weights may be negative.

Please find the negative weight cycles.


Input Format:

The first line contains three integers \( n \), \( m \), and \( k \).

The following \( m \) lines each contain three integers \( x \), \( y \), and \( z \), indicating that there is a directed edge from vertex \( x \) to vertex \( y \) with a length of \( z \).

Vertices are numbered from 1 to \( n \).


Output Format:

If there is a negative weight cycle in the graph, output `Yes`; otherwise, output `No`.




Constraints:

1 ≤ n ≤ 2000,

1 ≤ m ≤ 10000,

The absolute value of any edge weight does not exceed 10,000.



Sample Input:

3 3

1 2 -1

2 3 4

3 1 -4

Sample Output:

Yes

```python
from collections import deque
N,M = 100010, 100010
h,e,ne,w = [-1]*N, [-1]*N,[-1]*N,[0] * N
idx = 0
dist = [float('inf')] * N
count = [0] * N
dist[1] = 0

def add(a,b,z):
    global idx
    e[idx] = b
    ne[idx] = h[a]
    w[idx] = z
    h[a] = idx
    idx += 1

def spfa():
    visit = set()
    q = deque()
    for i in range(1,n+1):
        q.append(i)
        visit.add(i)
    for i in range(n):
        while len(q):
            point = q.popleft()
            visit.remove(point)
            next = h[point]
            while next != -1:
                j = e[next]
                if dist[j] > dist[point] + w[next]:
                    dist[j] = dist[point] + w[next]
                    count[j] = count[point] + 1
                    if count[j] >= n:
                        return 'Yes'
                    if j not in visit:
                        q.append(j)
                        visit.add(j)
                next = ne[next]
    return 'No'

n,m = map(int,input().split())
for i in range(m):
    a,b,z = map(int,input().split())
    add(a,b,z)

print(spfa())
```

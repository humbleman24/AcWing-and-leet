### 828 Stack
#### Problem Description

Initially, the stack is empty. The stack supports four types of operations:

1. `push x`: Insert a number x at the top of the stack.
2. `pop`: Remove and return the top element from the stack.
3. `empty`: Determine if the stack is empty.
4. `query`: Retrieve and output the top element of the stack.

You are required to perform M operations on the stack. For each operation of type 3 and type 4, output the corresponding result.


Input Format:

The first line contains an integer M, indicating the number of operations to be performed.

The following M lines each contain an operation command, which can be one of the following: push x, pop, empty, or query.

Output Format:

For each empty and query operation, output a corresponding query result, each result on a separate line.

For empty operations, the query result should be either YES or NO.

For query operations, the query result should be an integer representing the value of the top element of the stack.

Constraints:

1 ≤ M ≤ 100,000
1 ≤ x ≤ 10^9

All operations are guaranteed to be valid.

Sample Input:

10

push 5

query

push 6

pop

query

pop

empty

push 4

query

empty

Sample Output:

5

5

YES

4

NO

```python
N = 100010
arr = [0] * N
point = -1

M = int(input())

for i in range(M):
    command = input().split()
    if command[0] == 'push':
        point += 1
        arr[point] = command[1]
    elif command[0] == 'pop':
        arr.pop(point)
        point -= 1
    elif command[0] == 'empty':
        if point == -1:
            print('yes')
        else:
            print('no')
    elif command[0] == 'query':
        print(arr[point])
        
```

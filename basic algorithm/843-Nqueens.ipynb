### 843 N queens
#### Problem Description
The \( n \)-queens problem involves placing \( n \) queens on an \( n \times n \) chessboard in such a way that no two queens can attack each other. This means that no two queens can be in the same row, the same column, or on the same diagonal.


Input Format:

A single line containing one integer \( n \).


Output Format

Each solution should occupy \( n \) lines, with each line outputting a string of length \( n \) that represents the complete state of the chessboard.

Where '.' represents an empty square, and 'Q' represents a square occupied by a queen.

After completing the output of each solution, output an empty line.

Note: There should be no extra spaces at the end of each line.

The order of outputting the solutions can be arbitrary, as long as all are included and none are repeated.


Constraints

1 ≤ n ≤ 9


Sample Input:

4


Sample Output:

.Q..

...Q

Q...

..Q.
~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
..Q.

Q...

...Q

.Q..

```python
n = int(input())
matrix = [[1 for i in range(n)] for j in range(n)]
out = [['.' for i in range(n)] for j in range(n)]

def dfs(row):
    if row == n:
        for r in out:
            print(''.join(r))
        print()
        return None
    for index,cell in enumerate(matrix[row]):
        if cell >= 1:
            out[row][index] = 'Q'
            for r in range(row,n):
                matrix[r][index] -= 1
            r,c = row, index
            while r < n and c < n:
                matrix[r][c] -= 1
                r += 1
                c += 1
            r,c = row, index
            while r < n and c > -1:
                matrix[r][c] -= 1
                r += 1
                c -= 1
            
            dfs(row + 1)

            out[row][index] = '.'
            for r in range(row,n):
                matrix[r][index] += 1
            r,c = row, index
            while r < n and c < n:
                matrix[r][c] += 1
                r += 1
                c += 1
            r,c = row, index
            while r < n and c > -1:
                matrix[r][c] += 1
                r += 1
                c -= 1
    return None
dfs(0)
```

```python
# you don't need to keep track of all the cells!!! keep the index of column and diagonal and n-diagonal is enough

n = int(input())
out = [['.' for i in range(n)] for j in range(n)]
col , dg , ndg = [True] * n ,[True] * (2*n) ,[True] * (2*n)                          # represent whether that entire column or diagonal has element.

def dfs(row):
    if row == n:
        for r in out:
            print(''.join(r))
        print()
        return
    for i in range(n):
        if col[i] and dg[row+i] and ndg[n+i-row]:
            out[row][i] = 'Q'
            col[i] = dg[row+i] = ndg[n+i-row] = False
            dfs(row + 1)
            col[i] = dg[row+i] = ndg[n+i-row] = True
            out[row][i] = '.'
    return 
dfs(0)
```

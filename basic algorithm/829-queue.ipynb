### 828 Queue
#### Problem Description

Initially, the queue is empty. The queue supports four types of operations:

1. `push x`: Insert a number x at the rear of the queue.
2. `pop`: Remove and return the top element from the queue.
3. `empty`: Determine if the queue is empty.
4. `query`: Retrieve and output the top element of the queue.

You are required to perform M operations on the queue. For each operation of type 3 and type 4, output the corresponding result.


Input Format:

The first line contains an integer M, indicating the number of operations to be performed.

The following M lines each contain an operation command, which can be one of the following: push x, pop, empty, or query.

Output Format:

For each empty and query operation, output a corresponding query result, each result on a separate line.

For empty operations, the query result should be either YES or NO.

For query operations, the query result should be an integer representing the value of the top element of the queue.

Constraints:

1 ≤ M ≤ 100,000
1 ≤ x ≤ 10^9

All operations are guaranteed to be valid.

Sample Input:

10

push 6

empty

query

pop

empty

push 3

push 4

pop

query

push 6

Sample Output:

NO

6

YES

4

```python
N = 100010
arr = [0] * N
point = 0
first = -1

M = int(input())

for i in range(M):
    command = input().split()
    if command[0] == 'push':
        arr[point] = command[1]
        point += 1-;
    elif command[0] == 'pop':
        arr.pop(first + 1)
        first += 1
    elif command[0] == 'empty':
        if point - first == 1:
            print('yes')
        else:
            print('no')
    elif command[0] == 'query':
        print(arr[first+1])
        
```

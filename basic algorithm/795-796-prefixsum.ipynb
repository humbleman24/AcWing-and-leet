### 795 Prefix sum

#### Problem Description
Input a sequence of integers of length `n`.

Then input `m` queries, each query consists of a pair of integers `l, r`.

For each query, output the sum of numbers in the original sequence from the `l`-th number to the `r`-th number.

Input Format:

TThe first line contains two integers `n` and `m`.

The second line contains `n` integers, representing the sequence of integers.

The next `m` lines, each contain two integers `l` and `r`, representing the range of a query interval.


Output Format:

There are `m` lines in total, each line outputs the result of a query.


Constraints: 
1 ≤ `l` ≤ `r` ≤ `n`,
1 ≤ `n`, `m` ≤ 100000,
-1000 ≤ Value of elements in the sequence ≤ 1000


Sample Input:

5 3

2 1 3 6 4

1 2

1 3

2 4

Sample Output:

3

6

10


```python
# one dimension
n, m = list(map(int,input().split()))
a = list(map(int,input().split()))
accumulate = list()                   # calculate the Si ahead then just need one time iteration
accumulate.append(0)
for i in range(n):
    accumulate.append(accumulate[-1] + a[i])
i = 0
while i < m:
    i += 1
    s,r = list(map(int,input().split()))
    print(accumulate[r] - accumulate[s-1])
```

### 796 Prefix sum of matrix

#### Problem Description
Input an `n` by `m` matrix of integers, then input `q` queries, each query contains four integers `x1, y1, x2, y2`, representing the coordinates of the top-left corner and the bottom-right corner of a submatrix.

For each query, output the sum of all numbers in the submatrix.

Input Format:

The first line contains three integers `n`, `m`, `q`.

The next `n` lines, each contain `m` integers, representing the integer matrix.

The following `q` lines, each contain four integers `x1, y1, x2, y2`, representing a set of queries.


Output Format:

There are `m` lines in total, each line outputs the result of a query.


Constraints: 
1 ≤ `n`, `m` ≤ 1000,
1 ≤ `q` ≤ 200000,
1 ≤ `x1` ≤ `x2` ≤ `n`,
1 ≤ `y1` ≤ `y2` ≤ `m`,
-1000 ≤ Value of elements within the matrix ≤ 1000



Sample Input:

3 4 3

1 7 2 4

3 6 2 8

2 1 2 3

1 1 2 2

2 1 3 4

1 3 3 4

Sample Output:

17

27

21

```python
# two dimension
n,m,q = list(map(int,input().split()))
accumulate = list()
accumulate.append([0]*(m+1))                                   # m+1 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
for i in range(n):
    col = list(map(int,input().split()))
    acc = list()
    acc.append(0)
    for j in range(m):
        acc.append(acc[-1] + col[j])
    for j in range(m+1):
        acc[j] += accumulate[-1][j]
    accumulate.append(acc)

'''another way to calculate the accumulate by using the formula Sij = Si-1j + Sij-1 - Si-1j-1 + aij
accumulate = [[0 for i in range(m+1)] for j in range(n+1)]
for i in range(n):
    col = list(map(int,input().split()))
    for j in range(m):
        accumulate[i+1][j+1] = accumulate[i][j+1] + accumulate[i+1][j] - accumulate[i][j] + col[j]
'''
i = 0
while i < q:
    i += 1
    x1,y1,x2,y2 = list(map(int,input().split()))
    sum = accumulate[x2][y2] - accumulate[x1-1][y2] - accumulate[x2][y1-1] + accumulate[x1-1][y1-1]
    print(sum)
```

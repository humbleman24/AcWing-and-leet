### 860 dying to discriminate bipartite graph
#### Problem Description

Given an undirected graph with \( n \) vertices and \( m \) edges, the graph may contain parallel edges and self-loops.

Please determine whether this graph is a bipartite graph.

Input Format:

The first line contains two integers \( n \) and \( m \).

The following \( m \) lines each contain two integers \( u \) and \( v \), indicating that there is an edge between vertex \( u \) and vertex \( v \).



Output Format:

If the given graph is a bipartite graph, output `Yes`. Otherwise, output `No`.

Constraints:

1 ≤ n, m ≤ 10^5,


Sample Input:

4 4

1 3

1 4

2 3

2 4

Sample Output:

Yes

```python
from collections import deque
N = 100010
M = 2 * N
h, e, ne, color = [-1] * N, [0] * M, [0] * M, [0] * N
idx = 0

def add(a,b):
    global idx
    e[idx] = b
    ne[idx] = h[a]
    h[a] = idx
    idx += 1


def bfs():
    q = deque()
    for i in range(1, n+1):
        if not color[i]:
            q.appendleft((i,1))
            while len(q):
                p,c = q.pop()
                color[p] = c
                next = h[p]
                while next != -1:
                    j = e[next]
                    if not color[j]:
                        q.appendleft((j,-c))
                    elif color[j] == c:
                        return False
                    next = ne[next]
    return True

            

n,m = map(int,input().split())
for i in range(m):
    a,b = map(int,input().split())
    add(a,b)
    add(b,a)


if bfs(): print("Yes")
else: print("No")
```
